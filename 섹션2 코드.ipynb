## 결측치 처리하기

```python
df.count()
```

```python
df.isnull().sum()
```

```python
# 결측치가 100개 미만인 열 필터링
columns_to_keep = df.columns[df.isnull().sum() < 100]

# 결측치가 존재하는 행 삭제
df_cleaned = df.dropna(subset=columns_to_keep)
```

```python
df_cleaned.isnull().sum()
```

## 이상치(Outlier) 처리

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x=df_cleaned['Price'])
```

```python
df_cleaned[df_cleaned['Price'] >= 100000]
```

```python
df_cleaned.describe()
```

```python
pd.options.display.float_format = '{:.1f}'.format
```

```python
df_cleaned.describe()
```

```python
# IQR 계산
Q1 = df_cleaned['Price'].quantile(0.25)
Q3 = df_cleaned['Price'].quantile(0.75)
IQR = Q3 - Q1

# 아웃라이어 제거를 위한 상단/하단 경계 계산
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 아웃라이어 제거
df_filtered = df_cleaned[(df_cleaned['Price'] >= lower_bound) & (df_cleaned['Price'] <= upper_bound)]
```

```python
df_cleaned['Price'].quantile(0.25)
```

```python
df_filtered.shape
```

```python
sns.boxplot(x=df_filtered['Price'])
```

```python
import matplotlib.pyplot as plt

# 주어진 데이터
data = {
    'Norway': [5, 8],
    'Denmark': [4, 10],
    'Sweden': [13, 15]
}

years = [2004, 2022]

# 그래프 생성
fig, ax = plt.subplots()

# 각 국가의 막대 그래프 추가
for country, values in data.items():
    ax.bar(years, values, label=country)

# 그래프 스타일 설정
ax.set_xlabel('Year')
ax.set_ylabel('Value')
ax.set_title('Data Visualization')
ax.legend()

# 그래프 출력
plt.show()
```
